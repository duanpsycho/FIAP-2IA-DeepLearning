```
import keras
from keras.datasets import fashion_mnist
from matplotlib import pyplot
from random import randint
from keras.models import Sequential
```

```
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```
pyplot.imshow(x_train[randint(0, x_train.shape[0])], cmap='gray_r')
```

```
img_rows, img_cols = 28,28
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)
```

```
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```
from keras.layers import Conv2D, Dense, Flatten, MaxPooling2D

model = Sequential()
model.add(Conv2D(784, kernel_size=(3,3), activation='sigmoid', input_shape=input_shape))
model.add(Conv2D(78, kernel_size=(3,3), activation='sigmoid'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(78, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
training = model.fit(x_train, y_train, epochs=5, validation_split=0.1)
```
