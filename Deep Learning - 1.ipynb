```
# Importando datasets
from keras.datasets import mnist
```

```
# Separando entre treino e teste, Por padrão, 50%
(x_train, y_train), (x_test, y_test) = mnist.load_data() 
```

```
# Tamanho da Imagem
image_size = 784

# Transformação de Matriz para vetor, referente ao image_size
# Matriz sendo convertida para um vetor de 784 dimensoes
# Treino
x_train = x_train.reshape(x_train.shape[0], image_size) # Transformando matriz para vetor
x_train = x_train.astype('float32') # Como pontos flutuantes de 32bits
x_train /= 255 # Normalizando nossas entradas de 0-255 a 0.0-1.0

# Teste
x_test = x_test.reshape(x_test.shape[0], image_size) # Transformando matriz para vetor
x_test = x_test.astype('float32') # Como pontos flutuantes de 32bits
x_test /= 255 # Normalizando nossas entradas de 0-255 a 0.0-1.0
```

```
# Categorizando
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
```

```
# Modelo
# Referencia keras.oi/models/model
from keras.models import Sequential
```

```
# Iniciando o modelo
model = Sequential()

# Referencia keras.io/layers
from keras.layers import Dense
```

```
# Primeira layer
layer_input = Dense(units=512, activation = 'sigmoid', input_shape=(image_size,))

# Adicionando layer
model.add(layer_input)
```

```
# Hidden layer
model.add(Dense(units=512, activation='sigmoid'))
```

```
# Layer de saida
layer_out = Dense(units=10, activation='softmax')

# Adicionando a ultima camada
model.add(layer_out)
```

```
# Resumo
model.summary()
```

```
# Compilação

model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```
training = model.fit(x_train, y_train, epochs=20, validation_split=0.1)
```
